```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler,OneHotEncoder, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

Creamos la variable datos en la que utilizaremos la funcion read_csv de panda para leer el archivo Datos.csv de nuestro drive, quitaremos la columna 0 que esta de indice y especificaremos los tipos de varias columnas con dtype para que panda utilize el tipo especifico String en lugar del general Object que es el utilizado por defecto al leer strings ya que contiene al tipo string, pero queremos que sean strings para el tratado de los datos

```python
datos = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datos.csv', delimiter = ','
, skiprows= 0, dtype={'Sex': 'string','ChestPainType': 'string','RestingECG': 'string','ExerciseAngina': 'string','ST_Slope': 'string'})

datos
```

Por si acaso llamamos al metodo drop_duplicates() que borra cualquier fila duplicada del dataframe

```python
datos.drop_duplicates()
```

Creamos un nuevo DataFrame donde vamos a ordenar los valores por la columna de edad en orden descendiente con sort_values(by = ['Age'])

```python
datos_sorted = datos.sort_values(by = ['Age'])
datos_sorted
```

Como en datos.info() hemos podido observar que no tenemos valores nulos, vamos a crearlos artificialmente borrando datos aleatoriamente de la columna de Cholesterol que terminen en 3


```python
conteo = (datos['Cholesterol'] % 10 == 3).sum()
print("Cantidad de valores que terminan en 3:", conteo)
```

Ahora que sabemos que 83 filas contienen un dato Cholesterol que termina en 3 vamos a proceder a aplicar una funcion a toda la columna para que los borre

```python
datos_sorted['Cholesterol'] = datos_sorted['Cholesterol'].apply(lambda x: np.nan if x % 10 == 3 else x)
datos_sorted.loc[datos_sorted['Cholesterol'].isnull()]
```

Ahora tratamos esos datos nulos que hemos creado rellenandolos con 0 y corregimos el tipo ya que al meter nans en nuestra columna nos ha cambiado el tipo de datos a float para embarcar nan tambien, pero ahora que vamos a rellenar con 0s cambiamos el tipo de nuevo a int ya que los decimales son vacios

```python
datos_sorted.fillna(0, inplace=True)
datos_sorted['Cholesterol'] = datos_sorted['Cholesterol'].astype(int)
datos_sorted.info()
```

Hemos convertido a 0 los datos faltantes en lugar de dejarlos como nulos ya que en los datos originales podiamos encontrar algun valor 0 en la columna de colesterol asi que de esta manera para nuestro modelo de prediccion cogemos los valores borrados artificialmente y los ya faltantes de antemano

Ahora tratamos nuestros datos faltantes con un modelo de prediccion, hemos elegido el random forest regressor por su uso avanzado de arboles de decision para predecir el dato faltante basandose en otras columnas y otros valores de Cholesterol

Primero para utilizar el modelo hemos tenido que separar los datos en 2 dataframes, uno con las filas sin colesterol faltante y la otra solo con los faltantes

```python
df_known = datos_sorted[datos_sorted['Cholesterol'] != 0]
df_missing = datos_sorted[datos_sorted['Cholesterol'] == 0]
```

Una vez hecho eso vamos a convertir todas las columnas a numericas ya que el modelo requiere valores numeros para la prediccion, primero hacemos una rapida conversion de las columnas categoricas de cada dataframe con los datos faltantes y sin los datos faltantes y  despues unimos el resultado con las columnas ya numericas e comprobramos que los dos dataframes tengan el mismo indice para asegurarnos que no se ha perdido ninguna columna en el proceso

```python
features = ['Age', 'RestingBP', 'FastingBS', 'MaxHR', 'Oldpeak']
df_known_encoded = pd.get_dummies(df_known, columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
df_known_encoded[features] = df_known[features]
df_missing_encoded = pd.get_dummies(df_missing, columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
df_missing_encoded[features] = df_missing[features]
df_missing_encoded = df_missing_encoded.reindex(columns=df_known_encoded.columns, fill_value=0)
```

Ahora vamos a separar todo para el entrenamiento, por un lado vamos a coger las columnas con datos a usar como predictores "x_train" y por otra parte vamos a coger los datos de ejemplo de la columna a predecir "y_train". Despues metemos en un dataframe las columnas con datos vacios "x_missing"

```python
X_train = df_known_encoded.drop('Cholesterol', axis=1)
y_train = df_known_encoded['Cholesterol']
X_missing = df_missing_encoded.drop('Cholesterol', axis=1)
```

Por ultimo solo tenemos que crear el modelo de prediccion con 100 arboles y entrenar el modelo con los datos, una vez entrenado le metemos los datos que faltan para que los predizca y el resultado lo metemos en nuestra dataframe original en las posiciones correspondientes con valores 0 (convertimos los datos predecidos a ints ya que es el tipo original de la columna y para eso redondeamos y cambiamos el tipo a los datos predecidos antes de insertarlos en la dataframe original)

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
predicted_chol = model.predict(X_missing)
datos_sorted.loc[datos_sorted['Cholesterol'] == 0, 'Cholesterol'] = predicted_chol.round().astype(int)
datos_sorted.loc[datos_sorted['Cholesterol'].isin(predicted_chol.round().astype(int))]
```

Ahora vamos a hacer un proceso parecido pero con una columna categorica "ChestPainType" borrando el dato de esa columna en posiciones random, para eso vamos a crear una serie con numeros random dentro de nuestros limites para cambiar los valores en esas posiciones por NaN

Primero vamos a tener que transformar la columna de categorica a numerica con un encoder para que luego el modelo de prediccion se pueda utilizar ya que no se puede usar con columnas categoricas directamente

```python
pain_lbl = LabelEncoder()
datos_sorted['ChestPainType'] = pain_lbl.fit_transform(datos_sorted['ChestPainType'])
datos_sorted
```

```python
n_filas = int(len(datos_sorted)*0.35) #vamos a coger el 35% de los datos
np.random.seed(42)       # reproducible
random_indices = np.random.choice(datos_sorted.index, size=n_filas, replace=False)
datos_sorted.loc[random_indices, 'ChestPainType'] = np.nan
datos_sorted
```

El proceso es casi igual que con el Cholesterol, la unica diferencia es que ahora estamos utilizando los valores nulos directamente ya que en esta columna originalmente no hay ningun valor que no sea los cuatro tipos, vamos a hacer una numerizacion de las columnas categoricas y las uniremos a las ya numericas comprobando que el dataframe de datos conocidos y el dataframe de datos faltantes tienen el mismo indice

```python
datos_known = datos_sorted[datos_sorted['ChestPainType'].notna()]
datos_missing = datos_sorted[datos_sorted['ChestPainType'].isna()]
categoricas = ['Age','Cholesterol', 'RestingBP', 'FastingBS', 'MaxHR', 'Oldpeak']
datos_known_encoded = pd.get_dummies(datos_known, columns=['Sex', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
datos_known_encoded[categoricas] = datos_known[categoricas]
datos_missing_encoded = pd.get_dummies(datos_missing, columns=['Sex', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
datos_missing_encoded[categoricas] = datos_missing[categoricas]
datos_missing_encoded = datos_missing_encoded.reindex(columns=datos_known_encoded.columns, fill_value=0)
X_train = datos_known_encoded.drop('ChestPainType', axis=1)
y_train = datos_known_encoded['ChestPainType']
X_missing = datos_missing_encoded.drop('ChestPainType', axis=1)
```

Lo diferente esta vez es que en lugar de utilizar de modelo de prediccion randomforestregressor, vamos a utilizar otro algoritmo llamado xgboost que ,por ejemplo, utiliza arboles secuenciales en lugar de en paralelos como una de sus caracteristicas principales.

Primero declararemos el modelo y despues lo entrenaremos con los datos y por ultimo haremos que predizca los datos faltantes. Una vez hace todo eso, metemos los datos faltantes en la columna,al haber metido datos nulos anteriormente la columna ahora es de tipo float, antes de poder decodificarlos de vuelta a categorica cambiamos el tipo de vuelta a int y luego utilizamos la transformada inversa de nuestro codificador

```python
xgb = XGBClassifier(eval_metric='mlogloss', random_state=42)
xgb.fit(X_train, y_train)
predicted_type = xgb.predict(X_missing)
datos_sorted.loc[datos_sorted['ChestPainType'].isna(), 'ChestPainType'] = predicted_type
datos_sorted['ChestPainType'] = datos_sorted['ChestPainType'].astype(int)
datos_sorted['ChestPainType'] = pain_lbl.inverse_transform(datos_sorted['ChestPainType'])
datos_sorted.loc[random_indices]
```

Ahora vamos a comprobar como de bien ha predecido nuestro modelo los valores faltantes gracias a que tenemos los datos originales. Como podemos ver nos da un puntuaje de "0,61" entre 0 y 1, eso significa que el 60% de las veces nuestro modelo a acertado. Este puntuaje esta muy bien dado que estamos prediciendo un dato que no tiene correlacion directa con las otras columnas si no que depende de un dato psicologico real que da el paciente en el momento de la consulta

```python
y_true = datos.loc[random_indices, 'ChestPainType']
y_pred = datos_sorted.loc[random_indices, 'ChestPainType']
accuracy_score(y_true, y_pred)
```

Vamos a crear una nueva columna "CardioRisk" basandonos en el riesgo cardiaco calculado de cada fila sumando 5 tipicos indicadores de riesgo cardiaco con un 1 si esta fuera de los valores normales y 0 si esta dentro del rango normal

```python
datos_sorted['CardioRisk'] = (datos_sorted['RestingBP'] > 130).astype(int) + \
                   (datos_sorted['Cholesterol'] > 200).astype(int) + \
                   datos_sorted['FastingBS'] + \
                   (datos_sorted['Oldpeak'] > 2).astype(int) + \
                   (datos_sorted['ExerciseAngina'] == 'Y').astype(int)
datos_sorted
```
