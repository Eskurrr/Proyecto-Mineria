```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

Creamos la variable datos en la que utilizaremos la funcion read_csv de panda para leer el archivo Datos.csv de nuestro drive, quitaremos la columna 0 que esta de indice y especificaremos los tipos de varias columnas con dtype para que panda utilize el tipo especifico String en lugar del general Object que es el utilizado por defecto al leer strings ya que contiene al tipo string, pero queremos que sean strings para el tratado de los datos

```python
datos = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datos.csv', delimiter = ',', skiprows= 0, dtype={'Sex': 'string','ChestPainType': 'string','RestingECG': 'string','ExerciseAngina': 'string','ST_Slope': 'string'})

datos
```

Revisamos la correcta conversion a string de las columnas especificadas y de las demas columnas a tipos tratables

```python
datos.info()
```

Revisamos si el DataFrame contiene duplicados llamando al metodo sum en datos.duplicated y guardando el resultado en una variable para despues imprimir el resultado

```python
num_duplicates = datos.duplicated().sum()
print(f"El n√∫mero de filas duplicadas en el DataFrame es: {num_duplicates}")
```

Por si acaso llamamos al metodo drop_duplicates() que borra cualquier fila duplicada del dataframe

```python
datos.drop_duplicates()
```

Creamos un nuevo DataFrame donde vamos a ordenar los valores por la columna de edad en orden descendiente con sort_values(by = ['Age'])

```python
datos_sorted = datos.sort_values(by = ['Age'])
datos_sorted
```

Como en datos.info() hemos podido observar que no tenemos valores nulos, vamos a crearlos artificialmente borrando datos aleatoriamente de la columna de Cholesterol que terminen en 3


```python
conteo = (datos['Cholesterol'] % 10 == 3).sum()
print("Cantidad de valores que terminan en 3:", conteo)
```

Ahora que sabemos que 83 filas contienen un dato Cholesterol que termina en 3 vamos a proceder a aplicar una funcion a toda la columna para que los borre

```python
datos_sorted['Cholesterol'] = datos_sorted['Cholesterol'].apply(lambda x: np.nan if x % 10 == 3 else x)
datos_sorted.loc[datos_sorted['Cholesterol'].isnull()]
```

Ahora tratamos esos datos nulos que hemos creado rellenandolos con 0

```python
datos_sorted.fillna(0, inplace=True)
datos_sorted.info()
```

Vamos a crear una nueva columna "CardioRisk" basandonos en el riesgo cardiaco calculado de cada fila sumando 5 tipicos indicadores de riesgo cardiaco con un 1 si esta fuera de los valores normales y 0 si esta dentro del rango normal

```python
datos_sorted['CardioRisk'] = (datos_sorted['RestingBP'] > 130).astype(int) + \
                   (datos_sorted['Cholesterol'] > 200).astype(int) + \
                   datos_sorted['FastingBS'] + \
                   (datos_sorted['Oldpeak'] > 2).astype(int) + \
                   (datos_sorted['ExerciseAngina'] == 'Y').astype(int)
datos_sorted
```

Para la normalizacion primero tenemos que ver si tenemos valores atipicos para decidir si utilizar un minmaxscaler o un robustscaler ya que si normalizamos solo basandonos en minimos y maximos somos sensibles a los valores atipicos, utilizamos la funcion describe para sacar datos estadisticos de cada columna y ver el resultado

```python
datos_sorted.describe()
```

```python
datos_normalizados = datos_sorted.copy()
robust_scaler = RobustScaler()
datos_normalizados[['Age','RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak','CardioRisk']] = robust_scaler.fit_transform(datos_sorted[['Age','RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak','CardioRisk']])
datos_normalizados
```

```python
datos_sorted['ChestPainType'].unique()
```

```python
ata = datos_normalizados['ChestPainType'].apply(lambda x : 1 if x == 'ATA'else 0 )
ata.rename('ATA', inplace=True)
asy = datos_normalizados['ChestPainType'].apply(lambda x : 1 if x == 'ASY'else 0 )
asy.rename('ASY', inplace=True)
nap = datos_normalizados['ChestPainType'].apply(lambda x : 1 if x == 'NAP'else 0 )
nap.rename('NAP', inplace=True)
ta = datos_normalizados['ChestPainType'].apply(lambda x : 1 if x == 'TA'else 0 )
ta.rename('TA', inplace=True)
datos_normalizados = pd.concat([datos_normalizados.iloc[:,:2], ata, asy, nap, ta, datos_normalizados.iloc[:,3:]], axis=1)
datos_normalizados
```

```python
datos_sorted['Sex'].unique()
```

```python
datos_normalizados['Sex'] = datos_normalizados['Sex'].apply(lambda x : 1 if x == 'M'else 0 )
datos_normalizados
```

```python
datos_sorted['ExerciseAngina'].unique()
```

```python
datos_normalizados['ExerciseAngina'] = datos_normalizados['ExerciseAngina'].apply(lambda x : 1 if x == 'Y' else 0 )
datos_normalizados
```

```python
datos_sorted['ST_Slope'].unique()
```

```python
mapa = {'Up' : 1 , 'Flat':0.5 , 'Down':0}
datos_normalizados['ST_Slope'] = datos_normalizados['ST_Slope'].map(mapa)
datos_normalizados
```

```python
datos_normalizados['RestingECG'].unique()
```

```python
lvh = datos_normalizados['RestingECG'].apply(lambda x : 1 if x == 'LVH' else 0 )
lvh.rename('LVH', inplace=True)
normal = datos_normalizados['RestingECG'].apply(lambda x : 1 if x == 'Normal' else 0 )
normal.rename('Normal', inplace=True)
st = datos_normalizados['RestingECG'].apply(lambda x : 1 if x == 'ST' else 0 )
st.rename('ST', inplace=True)
datos_normalizados = pd.concat([datos_normalizados.iloc[:,:9], lvh, normal, st, datos_normalizados.iloc[:,10:]], axis=1)
datos_normalizados
```
