<div style="text-align: justify;">

  La insuficiencia cardíaca (IC), también conocida como insuficiencia cardíaca congestiva, es una afección grave que se produce cuando el músculo cardíaco no bombea la sangre como debería. El término "insuficiencia cardíaca" no significa que el corazón haya dejado de funcionar, sino que no es capaz de realizar el trabajo que se le exige para bombear suficiente sangre y satisfacer las demandas del organismo.
  
  Esta incapacidad puede deberse a que el corazón no se llena con suficiente sangre, o porque está demasiado débil o rígido para bombear adecuadamente. Cuando el corazón no bombea eficazmente, la sangre puede retroceder y provocar una acumulación de líquido (congestión).

  Las consecuencias de esta acumulación de líquido (edema) son evidentes en varios síntomas:
 * Acumulación en los pulmones, causando falta de aire.
 * Acumulación en otras partes del cuerpo, como hinchazón en las piernas, tobillos y pies.

  La insuficiencia cardíaca es a menudo causada por otras afecciones médicas que dañan o debilitan el corazón, como la enfermedad de las arterias coronarias (la causa más común) y la presión arterial alta.

  El presente proyecto busca abordar la predicción de esta enfermedad crítica utilizando un conjunto de datos robusto y curado y se titula "Conjunto de datos de predicción de insuficiencia cardíaca" [1].

  Este data set es notable porque fue creado al combinar cinco conjuntos de datos cardíacos independientes y previamente disponibles. La combinación de estos conjuntos (Cleveland, Húngaro, Suiza, Long Beach VA, y Stalog (corazón)) resultó en un total de 1190 observaciones, unificadas bajo 11 características comunes.

## 5.3 Tratar datos faltantes.
Para abordar los datos faltantes, se utilizará la columna de colesterol ya que es un dato bastante difícil de predecir, porque es un valor que viene directamente de pruebas de laboratorio, esto aumenta la complejidad del analisis. Se aplicarán dos métodos diferentes de predicción.

---
El primero será un enfoque "manual" que se basará en la media de un subgrupo, mientras que el segundo empleará un modelo de predicción más avanzado.

Dado que en `datos.info()` se ha observado que no hay valores nulos en ninguna de las columnas, entonces se crearan datos faltantes de forma artificial eliminando aleatoriamente algunos datos de la columna de colesterol.


1. Para el método manual, se hará una copia del DataFrame ordenada por edad, y se eliminarán los valores que terminen en siete para generar valores nulos "aleatorios".


---
Antes de realizar esta acción, se verificará cuántos datos se eliminarían.

```python
conteoM = (datos['Cholesterol'] % 10 == 7).sum()
print("Cantidad de valores que terminan en 7:", conteoM)
```

Después de confirmar que se eliminarán 71 valores, se creará una copia del DataFrame y se borraran esos datos.

```python
datos_null = datos_sorted.copy(deep=True)
datos_null['Cholesterol'] = datos_null['Cholesterol'].apply(lambda x: np.nan if x % 10 == 7 else x)
indicesnull = datos_null.index[datos_null['Cholesterol'].isnull()]
datos_null.loc[datos_null['Cholesterol'].isnull()]
```

A continuación, se calculará la media de colesterol agrupando los datos por edad y obteniendo la media correspondiente para cada grupo.


---
Una vez calculadas estas medias, se utilizará el método `fillna` para reemplazar cada valor faltante de manera respectiva.

```python
mean_cholesterol = datos_null.groupby('Age')['Cholesterol'].transform('mean')
datos_null['Cholesterol'].fillna(mean_cholesterol, inplace=True)
datos_null.loc[datos_null['Cholesterol'].isnull()]
```

```python
datos.loc[datos['Age'] == 30]
```

Se ha observado que existen un dato nulo debido a la ausencia de otros pacientes de 30 años que permitan calcular la media y completar el valor faltante de colesterol.

Dado que este dato es un valor de pruebas clinicas y no puede ser inferido a partir de otras columnas, se procederá a rellenarlo utilizando las medias para ese valor de los grupos de edad superior e inferior al paciente.

```python
mean_per_age = datos_null.groupby('Age')['Cholesterol'].mean()
valor_faltante = mean_per_age.loc[[29, 31]].mean()
datos_null.loc[datos_null['Cholesterol'].isnull(), 'Cholesterol'] = valor_faltante
datos_null.loc[datos_null['Cholesterol'] == valor_faltante]
```

Una vez que se han insertado las medias en los valores faltantes, se procederá a redondear y convertir todos los valores de la columna de colesterol a tipo entero, con el fin de regresar al formato original.

```python
datos_null['Cholesterol'] =  datos_null['Cholesterol'].round(0).astype(int)
datos_null
```

2. Para procedimiento automático se procederá a la eliminación de los registros cuyo valor finalice en 3.


---
Previamente, se realizará una verificación para determinar la cantidad exacta de datos que serían suprimidos.

```python
conteoA = (datos['Cholesterol'] % 10 == 3).sum()
print("Cantidad de valores que terminan en 3:", conteoA)
```

Una vez identificado que 83 filas contienen un valor en la variable Cholesterol que finaliza en 3, se procederá a aplicar una función sobre toda la columna con el fin de eliminar dichos registros.

```python
datos_sorted['Cholesterol'] = datos_sorted['Cholesterol'].apply(lambda x: np.nan if x % 10 == 3 else x)
datos_sorted.loc[datos_sorted['Cholesterol'].isnull()]
```

Los valores nulos generados en la columna se sustituyen por 0. Debido a la incorporación de NaN, el tipo de dato de la columna se transformó en float para permitir su representación. Sin embargo, al reemplazar los valores nulos por ceros, se procede a reconvertir la columna al tipo int, dado que los decimales carecen de significado en este contexto.

```python
datos_sorted.fillna(0, inplace=True)
datos_sorted['Cholesterol'] = datos_sorted['Cholesterol'].astype(int)
datos_sorted.info()
```

Los valores faltantes se sustituyeron por 0 en lugar de mantenerlos como nulos, dado que en los datos originales era posible encontrar registros con valor 0 en la variable Cholesterol. De este modo, el modelo de predicción considera tanto los valores eliminados artificialmente como los ausentes de forma inicial.

Posteriormente, el tratamiento de los datos faltantes se realizó mediante un modelo de predicción, seleccionándose el algoritmo Random Forest Regressor por su capacidad avanzada de emplear árboles de decisión para estimar los valores ausentes a partir de otras variables y de los registros existentes en la columna Cholesterol.

Para la implementación del modelo fue necesario dividir el conjunto de datos en dos dataframes: uno con las filas sin valores faltantes en colesterol y otro con las filas que presentaban dichos valores. Finalmente, se almacenaron los índices correspondientes a las columnas con datos ausentes.

```python
df_known = datos_sorted[datos_sorted['Cholesterol'] != 0]
df_missing = datos_sorted[datos_sorted['Cholesterol'] == 0]
indices0 = datos_sorted.index[datos_sorted['Cholesterol'] == 0]
```

Una vez completado este paso, todas las columnas se convierten a formato numérico, dado que el modelo de predicción requiere valores cuantitativos para su funcionamiento.


---
En primer lugar, se realiza una conversión rápida de las variables categóricas en ambos dataframes (*tanto el que contiene datos faltantes como el que no los presenta*).

Posteriormente, los resultados se integran con las columnas que ya eran numéricas y, finalmente, se verifica que ambos dataframes conserven el mismo índice, con el objetivo de garantizar que no se haya perdido ninguna variable durante el proceso.

```python
features = ['Age', 'RestingBP', 'FastingBS', 'MaxHR', 'Oldpeak']
df_known_encoded = pd.get_dummies(df_known, columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
df_known_encoded[features] = df_known[features]
df_missing_encoded = pd.get_dummies(df_missing, columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
df_missing_encoded[features] = df_missing[features]
df_missing_encoded = df_missing_encoded.reindex(columns=df_known_encoded.columns, fill_value=0)
```

En esta etapa se procede a la preparación de los datos para el entrenamiento del modelo.


---
Por un lado, se seleccionan las columnas que se utilizarán como variables predictoras `x_train`, mientras que, por otro, se extraen los valores de la columna objetivo que se desea predecir `y_train`.
Posteriormente, las columnas que presentan valores faltantes se agrupan en un dataframe independiente `x_missing`, con el fin de tratarlas de manera específica en el proceso de imputación.

```python
X_train = df_known_encoded.drop('Cholesterol', axis=1)
y_train = df_known_encoded['Cholesterol']
X_missing = df_missing_encoded.drop('Cholesterol', axis=1)
```

Finalmente, se construye el modelo de predicción utilizando un conjunto de 100 árboles y se entrena con los datos disponibles.


---

Una vez completado el entrenamiento, el modelo se aplica sobre los registros con valores faltantes para generar las predicciones correspondientes. Los resultados obtenidos se incorporan al dataframe original en las posiciones previamente ocupadas por ceros.

Con el fin de mantener la coherencia con el tipo de dato original de la columna, las predicciones se redondean y se convierten al formato int antes de su inserción.

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
predicted_chol = model.predict(X_missing)
datos_sorted.loc[datos_sorted['Cholesterol'] == 0, 'Cholesterol'] = predicted_chol.round().astype(int)
datos_sorted.loc[datos_sorted['Cholesterol'].isin(predicted_chol.round().astype(int))]
```

Se evaluó el desempeño de cada modelo mediante la comparación entre los valores predichos y los datos reales previamente eliminados de forma artificial.

---
Los resultados muestran que el modelo manual, basado en medias, no alcanzó ningún porcentaje de acierto, mientras que el modelo avanzado, fundamentado en algoritmos predictivos, obtuvo únicamente un 0,3% de precisión.

Este resultado es coherente, dado que el nivel de colesterol proviene de análisis clínicos reales y no de estimaciones estadísticas o cálculos derivados. No obstante, para el resto del código se opta por emplear el modelo avanzado, en virtud de su mayor score en la evaluación.

```python
y_trueM = datos.loc[indicesnull, 'Cholesterol']
y_predM = datos_null.loc[indicesnull, 'Cholesterol']
scoreM = accuracy_score(y_trueM, y_predM)
y_trueA = datos.loc[indices0, 'Cholesterol']
y_predA = datos_sorted.loc[indices0, 'Cholesterol']
scoreA = accuracy_score(y_trueA, y_predA)
print("Modelo manual score: ", (scoreM*100), "% , Modelo Avanzado: ", (scoreA*100), "%")
```

**3.2)** Se implementa un procedimiento análogo sobre la variable categórica `ChestPainType`.


---
Para ello, se eliminan de manera aleatoria determinados valores de la columna, generando posiciones con datos faltantes. Con este fin, se construye una serie de números aleatorios dentro de los límites establecidos, que se utiliza para sustituir los valores correspondientes por NaN.

En primer lugar, es necesario transformar la variable categórica en formato numérico mediante un encoder, dado que el modelo de predicción no puede operar directamente con columnas categóricas. Esta conversión garantiza la compatibilidad de los datos con el algoritmo y permite su correcta utilización en el proceso de modelado.

```python
pain_lbl = LabelEncoder()
datos_sorted['ChestPainType'] = pain_lbl.fit_transform(datos_sorted['ChestPainType'])
datos_sorted
```

Una vez realizada la conversión de la variable categórica a formato numérico, se procede a eliminar aleatoriamente un porcentaje de los datos mediante la función `np.random.choice.` Este enfoque se adopta en lugar del método previo de eliminación de registros que finalizaban en el dígito 3, ya que, al tratarse de un único dígito, resulta necesario emplear un procedimiento más avanzado para garantizar la aleatoriedad en el borrado.

```python
n_filas = int(len(datos_sorted)*0.35) #vamos a coger el 35% de los datos
np.random.seed(42)       # reproducible
random_indices = np.random.choice(datos_sorted.index, size=n_filas, replace=False)
datos_sorted.loc[random_indices, 'ChestPainType'] = np.nan
datos_sorted
```

El procedimiento es similar al aplicado en la variable *Cholesterol*, con la diferencia de que en este caso se emplean directamente los valores nulos, dado que la columna únicamente contiene los cuatro tipos definidos originalmente.

Se procede a la numerización de las variables categóricas y, posteriormente, se integran con las columnas ya numéricas. Finalmente, se verifica que tanto el dataframe con datos completos como el dataframe con valores faltantes mantengan el mismo índice, asegurando la coherencia estructural del conjunto de datos.

```python
datos_known = datos_sorted[datos_sorted['ChestPainType'].notna()]
datos_missing = datos_sorted[datos_sorted['ChestPainType'].isna()]
categoricas = ['Age','Cholesterol', 'RestingBP', 'FastingBS', 'MaxHR', 'Oldpeak']
datos_known_encoded = pd.get_dummies(datos_known, columns=['Sex', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
datos_known_encoded[categoricas] = datos_known[categoricas]
datos_missing_encoded = pd.get_dummies(datos_missing, columns=['Sex', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)
datos_missing_encoded[categoricas] = datos_missing[categoricas]
datos_missing_encoded = datos_missing_encoded.reindex(columns=datos_known_encoded.columns, fill_value=0)
X_train = datos_known_encoded.drop('ChestPainType', axis=1)
y_train = datos_known_encoded['ChestPainType']
X_missing = datos_missing_encoded.drop('ChestPainType', axis=1)
```

En esta ocasión, la diferencia metodológica radica en la elección del algoritmo de predicción.


---
En lugar de emplear `RandomForestRegressor`, se opta por utilizar `XGBoost`, cuya principal característica es la construcción de árboles de manera secuencial, a diferencia del enfoque en paralelo propio de los modelos basados en Random Forest. Esta estrategia permite optimizar el proceso de aprendizaje y mejorar la capacidad predictiva del modelo.

Se declara el modelo y posteriormente se entrena con los datos disponibles.


---
Una vez completado el entrenamiento, el modelo se aplica para predecir los valores faltantes. Las predicciones obtenidas se insertan en la columna correspondiente del dataframe.Debido a la incorporación previa de valores nulos, la columna se había transformado al tipo float; por ello, antes de proceder a la decodificación hacia su formato categórico original, se reconvierte al tipo int.

Finalmente, se aplica la transformación inversa del codificador para recuperar la representación categórica de la variable.

```python
xgb = XGBClassifier(eval_metric='mlogloss', random_state=42)
xgb.fit(X_train, y_train)
predicted_type = xgb.predict(X_missing)
datos_sorted.loc[datos_sorted['ChestPainType'].isna(), 'ChestPainType'] = predicted_type
datos_sorted['ChestPainType'] = datos_sorted['ChestPainType'].astype(int)
datos_sorted['ChestPainType'] = pain_lbl.inverse_transform(datos_sorted['ChestPainType'])
datos_sorted.loc[random_indices]
```

Esta clasificación se realizó distinguiendo entre **variables cualitativas** (nominales y ordinales) y **variables cuantitativas** (discretas y continuas). La diferenciación es esencial porque determina qué técnicas estadísticas y qué tipo de gráficos son más adecuados para representar cada atributo en el análisis exploratorio.  

- **Variables cualitativas nominales**: Son aquellas que representan categorías sin un orden específico. Por ejemplo, el sexo del paciente (M/F) o el resultado del electrocardiograma en reposo (Normal, ST, HVI). Estas variables permiten clasificar, pero no establecer jerarquías. - **Variables cualitativas ordinales**: También son categóricas, pero en este caso las categorías tienen un orden implícito. Un ejemplo es el tipo de dolor torácico (Angina típica, Angina atípica, Dolor no anginoso, Asintomático) o la pendiente del segmento ST (Ascendente, Plano, Descendente). El orden refleja una progresión clínica, aunque no se pueda cuantificar directamente. Se representan con gráficos de barras ordenadas o diagramas que reflejen la jerarquía.  

- **Variables cuantitativas discretas**: Son numéricas y toman valores enteros, generalmente contables. Ejemplos en el dataset son la edad, la presión arterial en reposo o el nivel de colesterol. Estas variables permiten calcular medidas de tendencia central y dispersión, y se suelen representar con histogramas o diagramas de caja.  

- **Variables cuantitativas continuas**: Son numéricas y pueden tomar cualquier valor dentro de un rango, incluyendo decimales. En el dataset, un ejemplo es la variable *Oldpeak* (depresión del segmento ST), que se mide en valores continuos. Estas variables permiten un análisis más detallado de la distribución y se representan con histogramas, densidades o diagramas de dispersión.

# Dispersión

---
La dispersión complementa el análisis al mostrar el grado de variabilidad existente en las observaciones. Con ella se puede evaluar hasta qué punto los valores se alejan de los centrales, lo que resulta clave para identificar heterogeneidad en el conjunto y posibles valores atípicos que podrían influir en la interpretación o en el rendimiento de los modelos predictivos.

Al estudiar variables clínicas cuantitativas como la presión arterial en reposo `RestingBP` o los niveles séricos de colesterol `Cholesterol`, no es suficiente describir únicamente su valor medio. Aunque la media permite obtener una primera aproximación del comportamiento general de la población, puede ocultar diferencias importantes entre los individuos. En el ámbito clínico, esta variabilidad es especialmente relevante, ya que dos poblaciones con la misma media pueden mostrar perfiles de riesgo completamente distintos.

Por ello, resulta fundamental calcular y analizar medidas de dispersión como la varianza y la desviación típica que indican en qué grado los valores se alejan de la media. Estas medidas permiten identificar si la variable está distribuida de manera homogénea o si, por el contrario, existe una amplitud considerable entre pacientes, lo cual podría reflejar condiciones clínicas heterogéneas, presencia de valores extremos o diferencias significativas en el estado cardiovascular.

En este análisis, la dispersión no solo sirve para describir mejor la distribución de los datos, sino también para interpretar adecuadamente la estabilidad, la consistencia y el riesgo asociado a cada variable.

## Hipótesis 1 para RestingBP:
---

Se plantea que la varianza y la desviación estándar de la presión arterial en reposo `RestingBP` reflejarán un grado significativo de dispersión en los valores respecto a la media, lo que indicaría la coexistencia de pacientes con presión arterial normal y otros con hipertensión persistente. Esta heterogeneidad es clínicamente relevante, porque según *Heart Failure Matters* la hipertensión arterial es una causa principal de insuficiencia cardíaca, al generar una sobrecarga crónica en el corazón que conduce a cambios estructurales y funcionales en el miocardio, aumentando el riesgo de fallo cardíaco. Por tanto, una varianza elevada en RestingBP podría evidenciar la presencia de distintos perfiles de riesgo cardiovascular dentro de la población estudiada.

```python
fig_restingbp = px.histogram(datos_sorted, x="RestingBP", title="Distribución de RestingBP con Desviación Estándar y Varianza", color_discrete_sequence=['rebeccapurple'])
fig_restingbp.update_layout(title_x=0.5, bargap=0.2, plot_bgcolor='thistle')

mean_restingbp = datos_sorted['RestingBP'].mean()
std_dev_restingbp = datos_sorted['RestingBP'].std()
variance_restingbp = datos_sorted['RestingBP'].var()

fig_restingbp.add_vline(x=mean_restingbp, line_dash="dash", line_width=2, line_color="hotpink", name="Media")
fig_restingbp.add_vline(x=mean_restingbp + std_dev_restingbp, line_dash="dot", line_width=2, line_color="mediumslateblue", name="+1 Std Dev")
fig_restingbp.add_vline(x=mean_restingbp - std_dev_restingbp, line_dash="dot", line_width=2, line_color="mediumslateblue", name="-1 Std Dev")

fig_restingbp.add_trace(go.Scatter(
    x=[None], y=[None],
    mode="lines",
    line=dict(color="hotpink", dash="dash"),
    name=f"Media = {mean_restingbp:.2f}"
))
fig_restingbp.add_trace(go.Scatter(
    x=[None], y=[None],
    mode="lines",
    line=dict(color="mediumslateblue", dash="dot"),
    name=f"Desv. Estándar = {std_dev_restingbp:.2f}"
))
fig_restingbp.add_trace(go.Scatter(
    x=[None], y=[None],
    mode="markers", # Use markers to show a single point in legend
    marker=dict(size=0, color="rgba(0,0,0,0)"), # Invisible marker
    name=f"Varianza = {variance_restingbp:.2f}"
))

fig_restingbp.show()
```

## Conclusión de la Hipotésis:
---

Los resultados muestran que la media de la presión arterial en reposo es elevada (132.4 mmHg), acompañada de una desviación estándar y varianza moderadas-altas (18.51 mmHg y 342.77, respectivamente), lo que indica una dispersión considerable de los valores. Esta variabilidad sugiere la existencia de diferentes perfiles de riesgo cardiovascular, desde pacientes normotensos hasta hipertensos. Dado que la hipertensión es un factor clave en el desarrollo de insuficiencia cardíaca —al provocar una sobrecarga crónica que daña el corazón y altera su función—, estos hallazgos subrayan la importancia de identificar y controlar la presión arterial elevada para prevenir la progresión hacia la insuficiencia cardíaca.
